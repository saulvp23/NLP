```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

delta = 0.5
x = np.arange(-5, 5, delta)
y = np.arange(-5, 5, delta)
X, Y = np.meshgrid(x, y)

Z = X**2 - Y**2 - X + Y -1

dX = 2*X -1
dY = -2*Y +1

fig, ax = plt.subplots()
fig.set_size_inches(10, 8, forward=True)
levels = np.arange(Z.min(),Z.max(),2)
norm = cm.colors.Normalize(vmax=abs(Z).max(), vmin=-abs(Z).max())

CS = ax.contourf(X, Y, Z, levels=levels, norm=norm, cmap=cm.coolwarm)
CS2 = ax.contour(X, Y, Z, levels=levels, norm=norm, colors='k')
fig.colorbar(CS, ax=ax)
ax.clabel(CS2, inline=1, fontsize=10)
ax.set_xlim([min(x), max(x)])
ax.set_ylim([min(y), max(y)])
ticksx = plt.xticks(np.arange(-5, 5, 0.5))
ticksy = plt.yticks(np.arange(-5, 5, 0.5))

q = ax.quiver(x, y, -dX, -dY)
```
